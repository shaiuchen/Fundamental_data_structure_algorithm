```python
import matplotlib.pyplot as plt
```

```python
# 讀取檔案並逐行處理
word_count = {}

with open('hw2_data.txt', 'r') as file:
    for line in file:
        word = line.strip().lower()  # 去除換行符號並轉成小寫
        if word:  # 確保不是空行
            word_count[word] = word_count.get(word, 0) + 1  # 使用 dict 累加單字次數

# 印出不重複的單字總數
print("不重複的英文單字數量：", len(word_count))

# 印出每個單字及其出現次數
print("\n每個英文單字出現次數：")
for word, count in word_count.items():
    print(f"{word}: {count}")

# 畫直方圖
sorted_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True) # 排序單字次數

# 分離單字與次數
words = [item[0] for item in sorted_words]
counts = [item[1] for item in sorted_words]

# 畫圖
plt.figure(figsize=(12, 6))
plt.bar(words, counts)
plt.xlabel('Words')
plt.ylabel('Times')
plt.tight_layout()  # 自動調整圖像排版以避免文字重疊
plt.show()
```
